## **My Second task is  Image to Pencil Sketch with Python.**


## **Import Module**

```python
import cv2
```

```python
image=cv2.imread("sai.jpg")
```

## **Create a New Image in GrayScale**

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

## **INVERTING OF GRAYSCALE IMAGE**

```python
inverted =255-gray_image
```

## **APPLY GAUSSIANBLUR()**

```python
blur = cv2.GaussianBlur(inverted,(21,21),0)
```

## **Inverting Of BlurImage**

```python
invertedblur =255- blur
```

## **IMAGE TO PENCIL SKETCH**

```python
sketch = cv2.divide(gray_image, invertedblur,scale=256.0)
```

## **Saving Of Image**

```python
cv2.imwrite("sketch.png",sketch)
```
